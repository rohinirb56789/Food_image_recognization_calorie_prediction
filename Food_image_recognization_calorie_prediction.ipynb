```python
# Food Recognition + Calorie Estimation — Colab-ready Notebook
# ----------------------------------------------------------
# Paste into a single Colab cell (or split into logical cells) and run top-to-bottom.
# No manual edits required (it will fallback to a synthetic dataset if TFDS/Kaggle isn't available).

# 0) Basic info / environment
import sys
print("Python:", sys.version)

# 1) Imports & configuration
import os
import random
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models

from sklearn.metrics import confusion_matrix, classification_report

# Reproducibility
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)

# Try to set GPU memory growth (safe)
gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus is not None and len(gpus) > 0:
    for g in gpus:
        try:
            tf.config.experimental.set_memory_growth(g, True)
        except Exception as ex:
            # If it fails, continue — not fatal
            print("Could not set memory growth:", ex)
else:
    print("No GPU found or could not query GPUs — running on CPU or default runtime.")

print("TensorFlow version:", tf.__version__)

# 2) Parameters
IMG_SIZE = (128, 128)           # input image size for the CNN
BATCH_SIZE = 32
AUTOTUNE = tf.data.AUTOTUNE
DEFAULT_NUM_CLASSES = 10       # we'll try to use up to this many classes from Food-101
EPOCHS = 12

# 3) Dataset handling
# Try TFDS Food-101 first. If fails, create a synthetic demo dataset.
use_synthetic = False
NUM_CLASSES = DEFAULT_NUM_CLASSES
class_names = None

try:
    import tensorflow_datasets as tfds
    print("Attempting to load Food-101 from tensorflow_datasets (this will download ~1GB).")
    # Load train + validation splits (as_supervised returns (image, label))
    ds_train_full, ds_val_full, ds_info = tfds.load(
        'food101',
        split=['train', 'validation'],
        with_info=True,
        as_supervised=True
    )
    all_class_names = ds_info.features['label'].names
    print(f"Food-101 total classes available: {len(all_class_names)}")
    # Use only the first NUM_CLASSES classes for speed (you can change DEFAULT_NUM_CLASSES)
    class_names = all_class_names[:NUM_CLASSES]
    print("Using classes:", class_names)

    # Keep only examples whose label < NUM_CLASSES
    def keep_first_n(img, label):
        return tf.less(label, NUM_CLASSES)

    ds_train = ds_train_full.filter(keep_first_n)
    ds_val = ds_val_full.filter(keep_first_n)

    # Optional: subsample to keep runtime friendly in Colab
    ds_train = ds_train.shuffle(20000, seed=SEED).take(4000)
    ds_val = ds_val.take(800)

    # Preprocessing function: resize and normalize, convert label to one-hot
    def preprocess(image, label):
        image = tf.image.resize(image, IMG_SIZE)
        image = tf.cast(image, tf.float32) / 255.0
        label = tf.one_hot(label, NUM_CLASSES)
        return image, label

    ds_train = ds_train.map(preprocess, num_parallel_calls=AUTOTUNE)
    ds_val = ds_val.map(preprocess, num_parallel_calls=AUTOTUNE)

    # Data augmentation (Keras preprocessing layers)
    data_augmentation = keras.Sequential([
        layers.RandomFlip('horizontal'),
        layers.RandomRotation(0.08),
        layers.RandomZoom(0.08),
    ], name='data_augmentation')

    ds_train = ds_train.batch(BATCH_SIZE).prefetch(AUTOTUNE)
    ds_val = ds_val.batch(BATCH_SIZE).prefetch(AUTOTUNE)

except Exception as e:
    print("Could not load Food-101 via TFDS (or TFDS not available). Falling back to synthetic dataset.")
    print("Reason:", e)
    use_synthetic = True

if use_synthetic:
    # Create a small synthetic dataset (two classes) to demonstrate the full pipeline.
    from PIL import Image, ImageDraw

    BASE_DIR = Path("synthetic_food_demo")
    train_dir = BASE_DIR / "train"
    val_dir = BASE_DIR / "val"
    classes = ['burger', 'salad']
    for cls in classes:
        (train_dir/cls).mkdir(parents=True, exist_ok=True)
        (val_dir/cls).mkdir(parents=True, exist_ok=True)

    # Create simple synthetic colored shapes as images for each class
    def make_synthetic_image(path, kind):
        img = Image.new("RGB", (200,200), color=(255,255,255))
        draw = ImageDraw.Draw(img)
        if kind == 'burger':
            # bun and greens
            draw.ellipse([30,80,170,150], fill=(194,109,47))
            draw.ellipse([40,95,160,125], fill=(34,139,34))
        else:  # salad
            draw.rectangle([50,50,150,150], fill=(34,139,34))
            # add some "toppings"
            draw.ellipse([70,70,90,90], fill=(255,215,0))
            draw.ellipse([110,90,125,105], fill=(255,0,0))
        img.save(path)

    # Generate images (small dataset to keep runtime low)
    for cls in classes:
        for i in range(200):  # train images
            make_synthetic_image(train_dir/cls/f"{cls}_{i}.jpg", cls)
        for i in range(50):   # val images
            make_synthetic_image(val_dir/cls/f"{cls}_{i}.jpg", cls)

    # Use keras utility to load datasets from directory
    ds_train = tf.keras.preprocessing.image_dataset_from_directory(
        str(train_dir), image_size=IMG_SIZE, batch_size=BATCH_SIZE, seed=SEED
    )
    ds_val = tf.keras.preprocessing.image_dataset_from_directory(
        str(val_dir), image_size=IMG_SIZE, batch_size=BATCH_SIZE, seed=SEED
    )
    class_names = ds_train.class_names
    NUM_CLASSES = len(class_names)
    print("Synthetic classes:", class_names)

    # Normalize and convert labels to one-hot
    normalization = layers.Rescaling(1./255)
    ds_train = ds_train.map(lambda x, y: (normalization(x), tf.one_hot(y, NUM_CLASSES)), num_parallel_calls=AUTOTUNE)
    ds_val   = ds_val.map(lambda x, y: (normalization(x), tf.one_hot(y, NUM_CLASSES)), num_parallel_calls=AUTOTUNE)

    # Augmentation (simpler)
    data_augmentation = keras.Sequential([
        layers.RandomFlip('horizontal'),
        layers.RandomRotation(0.08),
    ], name='data_augmentation')

# Ensure class_names and NUM_CLASSES are set
if class_names is None:
    class_names = [f"class_{i}" for i in range(NUM_CLASSES)]
print("Final NUM_CLASSES =", NUM_CLASSES)

# 4) Model building — CNN with BatchNorm and Dropout
def build_cnn(input_shape=IMG_SIZE + (3,), num_classes=NUM_CLASSES):
    inputs = keras.Input(shape=input_shape)
    x = data_augmentation(inputs)  # augmentation applied during training (layers handle training flag)

    x = layers.Conv2D(32, 3, padding='same', activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling2D()(x)

    x = layers.Conv2D(64, 3, padding='same', activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling2D()(x)

    x = layers.Conv2D(128, 3, padding='same', activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling2D()(x)

    x = layers.Flatten()(x)
    x = layers.Dropout(0.4)(x)
    x = layers.Dense(128, activation='relu')(x)
    x = layers.BatchNormalization()(x)
    outputs = layers.Dense(num_classes, activation='softmax')(x)

    model = keras.Model(inputs, outputs, name='food_cnn')
    return model

model = build_cnn()
model.summary()

# 5) Compile and train
model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=1e-3),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Callbacks — save the best model (Keras extension .keras)
BEST_PATH = "best_food_model.keras"
callbacks = [
    keras.callbacks.ModelCheckpoint(BEST_PATH, monitor='val_accuracy', save_best_only=True, verbose=1),
    keras.callbacks.EarlyStopping(monitor='val_accuracy', patience=4, restore_best_weights=True, verbose=1)
]

print("\nStarting training ...")
history = model.fit(
    ds_train,
    validation_data=ds_val,
    epochs=EPOCHS,
    callbacks=callbacks
)

# Save final model (explicit extension)
FINAL_MODEL_PATH = "final_food_model.keras"
model.save(FINAL_MODEL_PATH)
print("Final model saved to", FINAL_MODEL_PATH)

# 6) Evaluation: metrics, confusion matrix, classification report
# Convert validation dataset to numpy arrays (images and integer labels)
def dataset_to_numpy(ds):
    images = []
    labels = []
    # ds may contain one-hot labels in last axis
    for batch_x, batch_y in ds.unbatch().batch(256):
        images.append(batch_x.numpy())
        # If labels are one-hot convert to ints; if ints used already, handle both
        by = batch_y.numpy()
        if by.ndim > 1 and by.shape[-1] > 1:  # one-hot
            labels.append(np.argmax(by, axis=1))
        else:
            labels.append(by.reshape(-1))
    if len(images) == 0:
        return np.array([]), np.array([])
    X = np.concatenate(images, axis=0)
    y = np.concatenate(labels, axis=0)
    return X, y

X_val, y_val = dataset_to_numpy(ds_val)
print("Validation samples:", X_val.shape[0])

# Load best model and predict
best_model = keras.models.load_model(BEST_PATH)
y_prob = best_model.predict(X_val)
y_pred = np.argmax(y_prob, axis=1)

# Accuracy
acc = np.mean(y_pred == y_val)
print(f"\nValidation accuracy: {acc*100:.2f}%")

# Classification report
print("\nClassification report:")
print(classification_report(y_val, y_pred, target_names=class_names[:NUM_CLASSES]))

# Confusion matrix plot
cm = confusion_matrix(y_val, y_pred)
plt.figure(figsize=(8,6))
plt.imshow(cm, interpolation='nearest', cmap='viridis')
plt.title('Confusion Matrix')
plt.colorbar()
ticks = np.arange(min(NUM_CLASSES, len(class_names)))
plt.xticks(ticks, class_names[:NUM_CLASSES], rotation=45, ha='right')
plt.yticks(ticks, class_names[:NUM_CLASSES])
plt.tight_layout()
plt.show()

# 7) Plot training curves
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(history.history.get('accuracy', []), label='train_acc')
plt.plot(history.history.get('val_accuracy', []), label='val_acc')
plt.xlabel('Epoch'); plt.ylabel('Accuracy'); plt.legend()

plt.subplot(1,2,2)
plt.plot(history.history.get('loss', []), label='train_loss')
plt.plot(history.history.get('val_loss', []), label='val_loss')
plt.xlabel('Epoch'); plt.ylabel('Loss'); plt.legend()
plt.show()

# 8) Prediction demo — show some validation predictions and allow Colab upload
def show_predictions(model, images, max_images=5):
    if images.shape[0] == 0:
        print("No images to show.")
        return
    num = min(max_images, images.shape[0])
    probs = model.predict(images[:num])
    for i in range(num):
        plt.figure(figsize=(3,3))
        plt.imshow(images[i])
        lbl_idx = np.argmax(probs[i])
        plt.title(f"Pred: {class_names[lbl_idx]} ({probs[i][lbl_idx]:.2f})")
        plt.axis('off')
        plt.show()

print("\nSample validation predictions:")
if X_val.shape[0] > 0:
    show_predictions(best_model, X_val, max_images=5)
else:
    print("No validation images available to preview.")

# Optional: upload custom images when running inside Colab
try:
    from google.colab import files
    uploaded = files.upload()  # will open a file picker
    if uploaded:
        user_images = []
        from PIL import Image
        for fn in uploaded:
            img = Image.open(fn).convert('RGB').resize(IMG_SIZE)
            arr = np.array(img).astype('float32') / 255.0
            user_images.append(arr)
        user_images = np.stack(user_images, axis=0)
        print("\nPredictions for uploaded images:")
        show_predictions(best_model, user_images, max_images=len(user_images))
except Exception as e:
    # Not running in Colab or user cancelled upload - skip gracefully
    print("Upload step skipped (not running in Colab or upload cancelled).", e)

# 9) Toy calorie estimation heuristic
# NOTE: accurate calorie estimation requires segmentation + portion size + regression data.
# This is a simple deterministic mapping (for demo only).
print("\nToy calorie estimator (heuristic):")
calorie_map = {name: 200 + (abs(hash(name)) % 800) for name in class_names[:NUM_CLASSES]}
for name, cal in list(calorie_map.items())[:min(10, len(calorie_map))]:
    print(f"{name:20s} -> {cal} kcal")

# Demo: show predicted label + calorie for some validation images
if X_val.shape[0] > 0:
    probs = best_model.predict(X_val[:5])
    for i in range(min(5, X_val.shape[0])):
        idx = int(np.argmax(probs[i]))
        label = class_names[idx]
        est = calorie_map[label]
        print(f"Image #{i}: Predicted {label:20s} | Estimated calories: {est} kcal")

# END: Notebook tips
print("\nNotebook run complete.")
print("If you want a larger experiment, remove the .take(...) limits in the TFDS loading section and increase EPOCHS/IMG_SIZE.")
```
